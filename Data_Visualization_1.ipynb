# Line PLot
*Displays information as a series of datapoints called markers, connected by straight line segments

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'go--', linewidth=1, markersize=5)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Line PLot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
```

```python
x = np.arange(0,4 * np.pi, 0.1)#0.1 is step size
y = np.sin(x)
x
plt.title('Sine Waveform')

plt.plot(x,y,color = 'r')
plt.show()
```

```python
x=np.arange(0,10)
y= 3*x + 10
## plt plot
plt.plot(x,y,'ro--')
#plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.title('2d-Diagram', color='blue')
```

```python
#subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(0,10)
y = x*x
plt.title('Line PLot')
plt.xlabel('x-axis')
plt.plot(x,y,color = 'g')
plt.show()
```

```python
# Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

font1 = {'family':'serif','color':'blue','size':20}

plt.title('Sales Trend',color = 'b',fontdict = font1)
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.plot(days,sales_1,'r',label = 'Sales_1')
plt.plot(days,sales_2,'g',label = 'Sales_2')
plt.legend()#inside representation
plt.show()
```
